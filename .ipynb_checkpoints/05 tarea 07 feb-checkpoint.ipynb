<h3 style="color:#cc0066; margin-top:0;">üå∏ Pregunta 2 ‚Äî Sistemas de coordenadas (Astropy)</h3>

<p style="color:#1a1a1a; font-size:15px; margin:0 0 10px 0;">
<b>a) Altitud m√°xima (anal√≠tica)</b><br>
Para la culminaci√≥n (cuando el objeto pasa por el meridiano), la altitud m√°xima se estima con:
<br><br>
<span style="display:block; text-align:center;">
\( h_{\max} = 90^\circ - |\varphi - \delta| \)
</span>
<br>
En Bogot√°: \( \varphi = 4.65^\circ \) y para Betelgeuse \( \delta \approx +7.4^\circ \):
<br><br>
<span style="display:block; text-align:center;">
\( h_{\max} = 90^\circ - |4.65^\circ - 7.4^\circ| = 90^\circ - 2.75^\circ \approx 87.25^\circ \)
</span>
<br>
O sea, geom√©tricamente Betelgeuse puede culminar s√∫per alto. Si el enunciado dice que ‚Äúno supera 30¬∞‚Äù, eso normalmente se entiende como <i>no supera 30¬∞ durante la noche</i> (porque su culminaci√≥n puede caer de d√≠a).
</p>

<p style="color:#1a1a1a; font-size:15px; margin:0 0 10px 0;">
<b>b) ¬øPor qu√© Sgr A* se ve toda la noche pero Betelgeuse no?</b><br>
Porque eso depende de la fecha: por el movimiento de la Tierra alrededor del Sol, el cielo nocturno cambia durante el a√±o. En algunas √©pocas Sgr A* queda bien ‚Äúpuesto‚Äù en el cielo nocturno (separado del Sol) y dura muchas horas observable, mientras que Betelgeuse puede estar m√°s cerca del Sol en esa fecha o culminar en horas donde todav√≠a es de d√≠a, as√≠ que durante la noche no alcanza buena altitud.
</p>

<p style="color:#1a1a1a; font-size:15px; margin:0;">
<b>c) Medell√≠n (lat = 6.25¬∞N)</b><br>
Solo por geometr√≠a, la culminaci√≥n ser√≠a:
<br><br>
<span style="display:block; text-align:center;">
\( h_{\max} = 90^\circ - |6.25^\circ - 7.4^\circ| = 90^\circ - 1.15^\circ \approx 88.85^\circ \)
</span>

</p>

</div>

```python
import numpy as np
import astropy.units as u
from astropy.time import Time
from astropy.coordinates import SkyCoord, EarthLocation, AltAz, get_sun

# agregamos los datos relevantes 
date = "2026-02-05"
step = 1 * u.min  # resoluci√≥n: 1 minuto
times = Time(f"{date} 00:00:00", scale="utc") + np.arange(0, 24*60+1) * step  # 0..1440 min

# Betelgeuse 
betelgeuse = SkyCoord(ra=88.79*u.deg, dec=7.4*u.deg, frame="icrs")

def minutes_observable(lat, lon, height_m=2600):
    loc = EarthLocation(lat=lat*u.deg, lon=lon*u.deg, height=height_m*u.m)
    frame = AltAz(obstime=times, location=loc)

    alt_star = betelgeuse.transform_to(frame).alt
    alt_sun  = get_sun(times).transform_to(frame).alt

    night = alt_sun < (-18*u.deg)              # noche astron√≥mica
    visible = night & (alt_star > (30*u.deg))  # condici√≥n pedida

    minutes = int(np.sum(visible)) * 1  # porque step = 1 minuto
    hours = minutes / 60

    # m√°ximo de noche 
    max_alt_night = np.max(alt_star[night]).to(u.deg) if np.any(night) else np.nan*u.deg

    return minutes, hours, max_alt_night

# Bog y Mede (longitudes aprox)
bog_minutes, bog_hours, bog_max_night = minutes_observable(4.65, -74.07)
med_minutes, med_hours, med_max_night = minutes_observable(6.25, -75.58)

print("Bogot√°:")
print("  M√°xima altitud (solo de noche):", bog_max_night)
print("  Tiempo con alt > 30¬∞ y noche:", bog_minutes, "min =", round(bog_hours, 2), "h")

print("\nMedell√≠n:")
print("  M√°xima altitud (solo de noche):", med_max_night)
print("  Tiempo con alt > 30¬∞ y noche:", med_minutes, "min =", round(med_hours, 2), "h")

print("\n¬øBetelgeuse es observable > 2 horas?")
print("  Bogot√°  :", "S√≠" if bog_hours > 2 else "No")
print("  Medell√≠n:", "S√≠" if med_hours > 2 else "No")
```

<h3 style="color:#cc0066; margin-top:0;">üå∏ Pregunta 4 ‚Äî Interactive Analysis (Altair)</h3>

<p style="color:#1a1a1a; font-size:15px; margin:0 0 10px 0;">
<b>a) ¬øC√≥mo cambia el histograma de color al hacer brush en la secuencia principal (BP‚ÄìRP = 0.2 a 1.0, G &lt; 16)?</b><br>
El histograma se vuelve mucho m√°s enfocado: b√°sicamente se concentra solo en ese rango de color y desaparecen las colas extremas. 
Es como limpiar la muestra, porque estamos quitando estrellas que no pertenecen a la secuencia principal, como las muy rojas (gigantes) o muy azules. 
Entonces ya no se ve toda la poblaci√≥n mezclada, sino solo esa franja espec√≠fica.
</p>

<p style="color:#1a1a1a; font-size:15px; margin:0 0 10px 0;">
<b>b) ¬øPor qu√© el tama√±o del marcador se correlaciona con la paralaje?</b><br>
Porque la paralaje est√° directamente relacionada con qu√© tan cerca est√° una estrella. 
A mayor paralaje, la estrella est√° m√°s cerca (d ‚àù 1/œñ). 
Entonces, si el tama√±o depende de la paralaje, los puntos m√°s grandes representan estrellas cercanas y los peque√±os estrellas lejanas. 
Es una forma visual r√°pida de ver la distancia en el gr√°fico.
</p>

<p style="color:#1a1a1a; font-size:15px; margin:0;">
<b>c) Distribuci√≥n acumulada de magnitud absoluta</b><br>
Este punto se desarrolla en el c√≥digo de abajo.
</p>

</div>

```python
import numpy as np
import pandas as pd
import altair as alt


# DATOS 
np.random.seed(1997) #mi fecha de nacimiento
N = 1200
df = pd.DataFrame({
    "bp_rp": np.random.normal(1.0, 0.35, N),
    "phot_g_mean_mag": np.random.uniform(8, 18, N),
    "parallax": np.random.uniform(0.5, 15, N)  # mas
})

# Magnitud absoluta: M_G = m_G + 5*log10(parallax_mas) - 10
df["abs_mag"] = df["phot_g_mean_mag"] + 5*np.log10(df["parallax"]) - 10

# Selecci√≥n interactiva
brush = alt.selection_interval()

# Panel 1
scatter = (
    alt.Chart(df)
    .mark_circle(opacity=0.55)
    .encode(
        x=alt.X("bp_rp:Q", title="BP-RP"),
        y=alt.Y("abs_mag:Q", title="Magnitud absoluta", scale=alt.Scale(reverse=True)),
        size=alt.Size("parallax:Q", title="Paralaje (mas)"),
        color=alt.condition(brush, alt.value("steelblue"), alt.value("lightgray"))
    )
    .add_params(brush)
    .properties(width=280, height=250, title="Diagrama HR (selecci√≥n)")
)

# Panel 2
hist_color = (
    alt.Chart(df)
    .transform_filter(brush)
    .mark_bar(color="seagreen")
    .encode(
        x=alt.X("bp_rp:Q", bin=alt.Bin(maxbins=35), title="BP-RP"),
        y=alt.Y("count():Q", title="Conteo")
    )
    .properties(width=280, height=250, title="Histograma de color")
)

# Panel 3 (CDF)
cdf_mag = (
    alt.Chart(df)
    .transform_filter(brush)
    .transform_bin("M_bin", "abs_mag", bin=alt.Bin(maxbins=40))
    .transform_aggregate(n="count()", groupby=["M_bin"])
    .transform_window(
        cum_n="sum(n)",
        sort=[alt.SortField("M_bin", order="ascending")]
    )
    .transform_joinaggregate(total_n="sum(n)")
    .transform_calculate(cdf="datum.cum_n / datum.total_n")
    .mark_line(color="crimson", point=True)
    .encode(
        x=alt.X("M_bin:Q", title="Magnitud absoluta"),
        y=alt.Y("cdf:Q", title="Acumulada", scale=alt.Scale(domain=[0, 1]))
    )
    .properties(width=280, height=250, title="CDF magnitud absoluta")
)

final_chart = alt.hconcat(scatter, hist_color, cdf_mag)
final_chart
```

<h2 style="color:#ff4da6;">üíñ Pregunta 5 </h2>

<p><b style="color:#000000;">a) Pico ~15 min</b><br>
No coincide con los p-modes (~5 min). Es m√°s consistente con una oscilaci√≥n local de la regi√≥n activa, probablemente guiada por el campo magn√©tico (tipo magnetoac√∫stica de periodo largo).
</p>

<p><b style="color:#000000;">b) Submap vs disco completo</b><br>
El promedio del disco mezcla muchas se√±ales y diluye la de inter√©s. El submap a√≠sla la regi√≥n activa, mejora la se√±al y reduce el ruido de otras zonas.
</p>

<p><b style="color:#000000;">c) Ajustar seno amortiguado y comparar</b><br>
En el c√≥digo de abajo se puede observar el ajuste de un seno amortiguado a <i>times</i> y <i>light_curve</i>, y la comparaci√≥n mediante <span style="color:#ff4da6;">$\chi^2$</span> para evaluar qu√© tan bien describe la variaci√≥n observada.
</p>

</div>

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit


# mi semilla 
np.random.seed(1997)


# 2) Simular lightcurve (porque no hay FITS)

t = np.linspace(0, 100, 200)

# se√±al tipo solar oscilante amortiguada + ruido
y = 2 * np.exp(-t/50) * np.sin(2*np.pi*t/15) + 0.3*np.random.normal(size=len(t))


# 3) Modelo
def damped_sine(t, A, P, tau, phi, C):
    return A * np.exp(-t/tau) * np.sin(2*np.pi*t/P + phi) + C

# guesses
p0 = [2, 15, 50, 0, 0]


# 4) Fit

popt, _ = curve_fit(damped_sine, t, y, p0=p0)
y_fit = damped_sine(t, *popt)


# 5) chi^2 (x2)

sigma = 0.3
chi2 = np.sum(((y - y_fit)/sigma)**2)
dof = len(y) - len(popt)
chi2_red = chi2 / dof

print("Par√°metros:", popt)
print("chi^2 =", chi2)
print("chi^2 reducido =", chi2_red)


# 6) Plot
plt.figure(figsize=(10,4))
plt.plot(t, y, "o", ms=4, label="Lightcurve")
plt.plot(t, y_fit, "-", lw=3, label="Fit sinusoide amortiguada")
plt.legend()
plt.grid(alpha=0.3)
plt.title("Fit sinusoide amortiguada (simulado)")
plt.show()
```
