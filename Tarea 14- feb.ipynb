## ðŸŒ¸ Derivadas Centradas en el Punto Medio

En el punto medio $D = 14 \,\text{Mpc}$ usamos fÃ³rmulas centradas porque tenemos informaciÃ³n a ambos lados del punto.

Datos:
- $z_1 = 0.0035$
- $z_2 = 0.0055$
- $z_3 = 0.0080$
- $h = 2 \,\text{Mpc}$

---

### ðŸ’– Primera derivada

FÃ³rmula centrada:

$$
\frac{dz}{dD} = \frac{z_{i+1} - z_{i-1}}{2h}
$$

Sustituyendo:

$$
\frac{dz}{dD} = \frac{0.0080 - 0.0035}{2(2)}
$$

$$
\frac{dz}{dD} = 0.001125 \,\text{Mpc}^{-1}
$$

---

### ðŸ’— Segunda derivada

FÃ³rmula centrada de segundo orden:

$$
\frac{d^2z}{dD^2} = \frac{z_{i+1} - 2z_i + z_{i-1}}{h^2}
$$

Sustituyendo:

$$
\frac{d^2z}{dD^2} = \frac{0.0080 - 2(0.0055) + 0.0035}{2^2}
$$

$$
\frac{d^2z}{dD^2} = 0.000125 \,\text{Mpc}^{-2}
$$

---

## ðŸŒ· ConclusiÃ³n

En $D=14\,\mathrm{Mpc}$ obtuvimos:

$$
\frac{dz}{dD}\approx 0.001125\ \mathrm{Mpc}^{-1},
\qquad
\frac{d^2z}{dD^2}\approx 0.000125\ \mathrm{Mpc}^{-2}.
$$

El hecho de que $\dfrac{dz}{dD}>0$ indica que el redshift $z$ aumenta con la distancia $D$, consistente con la expansiÃ³n.

Usamos fÃ³rmulas **centradas** porque estamos en un punto interior (tenemos datos a ambos lados).  
AdemÃ¡s, estas aproximaciones son de orden $O(h^2)$, por lo que el error de truncamiento decrece proporcionalmente a $h^2$, siendo mÃ¡s precisas que un esquema de primer orden $O(h)$.

```python
# Ejercicio 5: Error Analysis and Consistency Discussion
# Pregunta 2: Convergence and Step Size (p = 2)
p = 2
h = 0.1
C = 1.0

E_h = C * h**p
E_half = C * (h/2)**p

factor = E_half / E_h

print(f"E(h)      = {E_h:.6f}")
print(f"E(h/2)    = {E_half:.6f}")
print(f"E(h/2)/E(h) = {factor:.6f}")
print(f"Reduction factor = {1/factor:.0f}  (i.e., error is 4x smaller)")




```

ðŸ©· <b>â‘  Theoretical Accuracy</b>  

El mÃ©todo centrado es mÃ¡s confiable que el forward porque su error de truncamiento es $O(h^2)$ y no $O(h)$.  
En el esquema centrado se cancelan tÃ©rminos de Taylor de primer orden, dejando como error dominante un tÃ©rmino proporcional a $h^2$.

<hr style="border:1px solid #ffb6c1;">

ðŸ©· <b>â‘¡ Convergence and Step Size</b>  

Para un mÃ©todo de segundo orden ($p=2$), el error escala como:

$$
E(h) \sim h^2
$$

Si el paso se reduce a la mitad:

$$
E\left(\frac{h}{2}\right)
=
\left(\frac{1}{2}\right)^2 E(h)
=
\frac{1}{4}E(h)
$$

Por tanto, el error disminuye por un factor de 4.
